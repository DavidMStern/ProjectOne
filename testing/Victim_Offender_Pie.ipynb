```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file = "../clean_data/clean_victimization_data.csv"
df = pd.read_csv(data_file)
pd.set_option('display.max_columns', 50)
#df.head(50)
```

```python
df_vi = df['Victim-offender relationship'].value_counts(sort=False)
df_vi
```

```python
#pie chart to show victim-offender relationship

colors = ["red", "blue", "green", "yellow", "purple", "orange"]
crimes = df["Victim-offender relationship"].value_counts(sort=False)
plt.pie(crimes, labels=df_vi.index, colors=colors, autopct="%1.1f%%", shadow=True, startangle=180)

plt.savefig("victim_offender_pie")
plt.show()

```
