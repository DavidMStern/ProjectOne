```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
pd.set_option("display.max_columns", 50)
```

```python
# Making dataframe from clean data
df = pd.read_csv("../clean_data/clean_victimization_data.csv")
df.head()
```

```python
df['Household income'].value_counts()
```

```python
graph_df = pd.DataFrame()

for age in df['Age'].unique():
    print(age)
    age_df = df.copy()[df['Age'] == age]
    age_df[f"{age}"] = age_df['Household income'].value_counts().index.tolist()
    
    if graph_df.empty:
        graph_df = age_df[[f"{age}"]]
    else:
        graph_df = graph_df.join(age_df[f"{age}"])
```
