```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('../clean_data/clean_victimization_data.csv')
df.head(10)
```

```python
RCI = (df.pivot_table(index='Type of crime',
                      columns='Race',
                      values= df.groupby('Household income').count(),
                      aggfunc='sum'))
RCI.plot(kind="bar", stacked=False)
plt.title('Crimes based on Income by Race')
plt.ylabel('Household Income')
plt.legend(df['Race'])
plt.savefig("Crimes by race by income")
plt.show()
```

```python
RSA = (df.pivot_table(index='Type of crime',
                      columns='Sex',
                      values= df.groupby('Age').count(),
                      aggfunc='sum'))
RSA.plot(kind="bar", stacked=False)
plt.title('Types of Crime based on Age by Gender')
plt.ylabel('Age')
plt.legend(df['Sex'])
plt.savefig("CrimeAgeGender")
plt.show()
```
