```python
import pandas as pd
```

```python
df_1 = pd.read_csv('data_one.csv')
df_1.head()
```

```python
d_gender = {1: "Male", 2: "Female"}
df_1 = df_1.replace({"gender": d_gender})

d_race = {1: "White", 2: "Black", 3: "Other"}
df_1 = df_1.replace({"race1r": d_race})

d_hispanic = {1: "Hispanic", 2: "Non-Hispanic"}
df_1 = df_1.replace({"hispanic": d_hispanic})

d_ethnic = {1: "Non-Hispanic White", 2: "Non-Hispanic Black", 3: "Non-Hispanic", 4: "Other"}
df_1 = df_1.replace({"ethnic1r": d_ethnic})

d_age = {1: "12-14", 2: "15-17", 3: "18-20", 4: "21-24", 5: "25-34", 6: "35-49", 7: "50-64", 8: "65+"}
df_1 = df_1.replace({"ager": d_age})

d_marital = {1: "Never married", 2: "Married", 3: "Widowed", 4: "Divorced", 5: "Separated"}
df_1 = df_1.replace({"marital2": d_marital})

df_1.head()
```

```python
df_1.to_csv("data_one_fix.csv")
```
