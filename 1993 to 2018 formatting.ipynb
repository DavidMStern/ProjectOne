```python
import pandas as pd
pd.set_option('display.max_columns', 50)
```

```python
victimization_data = pd.read_csv('NCVS_PERSONAL_1993-2018/NCVS_PERSONAL_VICTIMIZATION_1993-2018.csv')
victimization_data
```

```python
victimization_data['locationr'].value_counts()
```

```python
victimization_data['gender'] = victimization_data['gender'].replace({1:'Male', 2:'Female'})
victimization_data['race1r'] = victimization_data['race1r'].replace({1:'White', 2:'Black', 3:'Other'})
victimization_data['hispanic'] = victimization_data['hispanic'].replace({1:'Hispanic', 2:'Non-Hispanic', 88:'Unknown'})
victimization_data['ethnic1r'] = victimization_data['ethnic1r'].replace({1:'Non-Hispanic White', 2:'Non-Hispanic Black', 
                                                                         3:'Non-Hispanic Other', 4:'Hispanic'})
victimization_data['ager'] = victimization_data['ager'].replace({1:'12 to 14', 2:'15 to 17', 3:'18 to 20', 4:'21 to 24', 
                                                                 5:'25 to 34', 6:'35 to 49', 7:'50 to 64', 8:'65 and older'})
victimization_data['marital2'] = victimization_data['marital2'].replace({1:'Never Married', 2:'Married', 3:'Widowed', 
                                                                         4:'Divorced', 5:'Separated', 88:'Unknown'})
victimization_data['hincome'] = victimization_data['hincome'].replace({1:'Less than $7,500', 2:'7,500 to 14,999', 
                                                                       3:'15,000 to 24,999', 4:'25,000 to 34,999', 
                                                                       5:'35,000 to 49,999', 6:'50,000 to 74,999', 
                                                                       7:'75,000 or more', 88:'Unknown'})
victimization_data['popsize'] = victimization_data['popsize'].replace({0.0:'Not a place', 1.0:'Under 100,000', 
                                                                       2.0:'100,000 to 249,999', 3.0:'250,000 to 499,999', 
                                                                       4.0:'500,000 to 999,999', 5.0:'1 million or more'})
victimization_data['region'] = victimization_data['region'].replace({1:'Northeast', 2:'Midwest', 3:'South', 
                                                                     4:'West', -1:'Unknown'})
victimization_data['msa'] = victimization_data['msa'].replace({1:'Urban', 2:'Suburban', 3:'Rural'})
victimization_data['direl'] = victimization_data['direl'].replace({1:'Intimates', 2:'Other Relatives',
                                                                   3:'Well-known/Casual Acquaintances', 4:'Stanger', 
                                                                   5:'Do not know relationship', 
                                                                   6:'Unknown number of offenders'})
victimization_data['notify'] = victimization_data['notify'].replace({1:'Yes', 2:'No', 3:'Unknown', 8:'Unknown'})
victimization_data['weapon'] = victimization_data['weapon'].replace({1:'Yes', 2:'No', 3:'Unknown'})
victimization_data['weapcat'] = victimization_data['weapcat'].replace({0:'No Weapon', 1:'Firearm', 2:'Knife', 3:'Other', 
                                                                       4:'Type Unknown', 5:'Unknown if offender had weapon'})
victimization_data['newcrime'] = victimization_data['newcrime'].replace({1:'Violent Victimization', 2:'Personal Theft/Larceny'})
victimization_data['newoff'] = victimization_data['newoff'].replace({1:'Rape/Sexual Assaul', 2:'Robbery', 
                                                                     3:'Aggravated Assault', 4:'Simple Assault', 
                                                                     5:'Personal Theft'})
victimization_data['seriousviolent'] = victimization_data['seriousviolent'].replace({1:'Violent Crime', 2:'Simple Assault', 
                                                                                     3:'Personal Theft', 4:'Property Crime'})
victimization_data['injury'] = victimization_data['injury'].replace({0:'Not Injured', 1:'Injured'})
victimization_data['treatment'] = victimization_data['treatment'].replace({0.0:'Not Injured', 1.0:'Not Treated', 2.0:'Treated'})
victimization_data['vicservices'] = victimization_data['vicservices'].replace({1.0:'Service Received', 
                                                                               2.0:'No Services Recieved'})
victimization_data['locationr'] = victimization_data['locationr'].replace({1:'Home', 2:'Friendly Location', 3:'Commercial', 
                                                                           4:'School', 5:'Other'})
victimization_data.head()
```

```python
victimization_data = victimization_data.rename(columns={'year':'Year', 'weight':'Weight', 'gender':'Gender', 'race1r':'Race', 
                                   'hispanic':'Hispanic Origin', 'ethnic1r':'Race/Hispanic Origin', 'ager':'Age', 
                                   'marital2':'Marital Status', 'hincome':'Household Income', 'popsize':'Population Size', 
                                   'region':'Region', 'msa':'Location of Residence', 'direl':'Victim-offender Relationship', 
                                   'notify':'Police Report', 'weapon':'Presence of Weapon', 'weapcat':'Weapon Category', 
                                   'newcrime':'Aggregate Type of Crime', 'newoff':'Type of Crime', 
                                   'seriousviolent':'Violent Crime Excluding Simple Assault', 'injury':'Injury', 
                                   'treatment':'Treatment', 'vicservices':'Victim Services', 'locationr':'Location of Incident'})
victimization_data.head()
```

```python
victimization_data = victimization_data.dropna(how='any')
victimization_data.head()
```

```python
victimization_data.to_csv('clean_data/clean_victimization_data.csv', index=False)
```
