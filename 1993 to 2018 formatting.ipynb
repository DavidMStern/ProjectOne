```python
import pandas as pd
pd.set_option('display.max_columns', 50)
```

```python
victimization_data = pd.read_csv('NCVS_PERSONAL_1993-2018/NCVS_PERSONAL_VICTIMIZATION_1993-2018.csv')
victimization_data
```

```python
victimization_data['locationr'].value_counts()
```

```python
victimization_data['gender'] = victimization_data['gender'].replace({1:'Male', 2:'Female'})
victimization_data['race1r'] = victimization_data['race1r'].replace({1:'White', 2:'Black', 3:'Other'})
victimization_data['hispanic'] = victimization_data['hispanic'].replace({1:'Hispanic', 2:'Non-Hispanic', 88:'Unknown'})
victimization_data['ethnic1r'] = victimization_data['ethnic1r'].replace({1:'Non-Hispanic white', 2:'Non-Hispanic black', 
                                                                         3:'Non-Hispanic other', 4:'Hispanic'})
victimization_data['ager'] = victimization_data['ager'].replace({1:'12 to 14', 2:'15 to 17', 3:'18 to 20', 4:'21 to 24', 
                                                                 5:'25 to 34', 6:'35 to 49', 7:'50 to 64', 8:'65 or older'})
victimization_data['marital2'] = victimization_data['marital2'].replace({1:'Never married', 2:'Married', 3:'Widowed', 
                                                                         4:'Divorced', 5:'Separated', 88:'Unknown'})
victimization_data['hincome'] = victimization_data['hincome'].replace({1:"Less than $7,500", 2:"$7,500 to $14,999", 
                                                                       3:"$15,000 to $24,999", 4:"$25,000 to $34,999", 
                                                                       5:"$35,000 to $49,999", 6:"$50,000 to $74,999", 
                                                                       7:"$75,000 or more", 88:"Unknown"})
victimization_data['popsize'] = victimization_data['popsize'].replace({0.0:'Not a place', 1.0:'Under 100,000', 
                                                                       2.0:'100,000 to 249,999', 3.0:'250,000 to 499,999', 
                                                                       4.0:'500,000 to 999,999', 5.0:'1 million or more'})
victimization_data['region'] = victimization_data['region'].replace({1:'Northeast', 2:'Midwest', 3:'South', 
                                                                     4:'West', -1:'Unknown'})
victimization_data['msa'] = victimization_data['msa'].replace({1:'Urban', 2:'Suburban', 3:'Rural'})
victimization_data['direl'] = victimization_data['direl'].replace({1:'Intimates', 2:'Other relatives',
                                                                   3:'Well-known/casual acquaintances', 4:'Stanger', 
                                                                   5:'Do not know relationship', 
                                                                   6:'Do not know number of offenders'})
victimization_data['notify'] = victimization_data['notify'].replace({1:'Yes, reported to the police',
                                                                     2:'No, did not report to the police', 3:'Do not know',
                                                                     8:'Unknown'})
victimization_data['weapon'] = victimization_data['weapon'].replace({1:'Yes, offender had weapon',
                                                                     2:'No, offender did not have weapon',
                                                                     3:'Do not know if offender had weapon'})
victimization_data['weapcat'] = victimization_data['weapcat'].replace({0:'No weapon', 1:'Firearm', 2:'Knife',
                                                                       3:'Other type weapon', 4:'Type weapon unknown',
                                                                       5:'Do not know if offender had weapon'})
victimization_data['newcrime'] = victimization_data['newcrime'].replace({1:'Violent victimization', 2:'Personal theft/larceny'})
victimization_data['newoff'] = victimization_data['newoff'].replace({1:'Rape/sexual assault', 2:'Robbery', 
                                                                     3:'Aggravated assault', 4:'Simple assault', 
                                                                     5:'Personal theft'})
victimization_data['seriousviolent'] = victimization_data['seriousviolent'].replace({1:'Violent crime excluding simple assault',
                                                                                     2:'Simple assault', 3:'Personal theft',
                                                                                     4:'Property crime'})
victimization_data['injury'] = victimization_data['injury'].replace({0:'Not injured', 1:'Injured'})
victimization_data['treatment'] = victimization_data['treatment'].replace({0.0:'Not injured', 1.0:'Not treated for injury',
                                                                           2.0:'Treated at scene, home, medical office, or other location'})
victimization_data['vicservices'] = victimization_data['vicservices'].replace({1.0:'Services received from victim service agencies', 
                                                                               2.0:'No services received from victim service agencies'})
victimization_data['locationr'] = victimization_data['locationr'].replace({1:"At or near victim's home",
                                                                           2:"At or near friend, neighbor, or relative's home",
                                                                           3:'Commercial place, parking lot, or other public area', 
                                                                           4:'School', 5:'Other location'})
victimization_data.head()
```

```python
victimization_data = victimization_data.rename(columns={'year':'Year', 'weight':'Weight', 'gender':'Sex', 'race1r':'Race', 
                                                        'hispanic':'Hispanic origin', 'ethnic1r':'Race/Hispanic origin', 
                                                        'ager':'Age', 'marital2':'Marital status', 'hincome':'Household income',
                                                        'popsize':'Population size', 'region':'Region',
                                                        'msa':'Location of residence', 'direl':'Victim-offender relationship',
                                                        'notify':'Reporting to the police', 'weapon':'Presence of Weapon',
                                                        'weapcat':'Presence of weapon', 'newcrime':'Aggregate type of crime',
                                                        'newoff':'Type of crime', 'seriousviolent':'Violent crime excluding simple assault',
                                                        'injury':'Injury', 'treatment':'Medical treatment for physical injuries',
                                                        'vicservices':'Victim services', 'locationr':'Location of incident'})
victimization_data.head()
```

```python
victimization_data = victimization_data.dropna(how='any')
victimization_data.head()
```

```python
victimization_data.to_csv('clean_data/clean_victimization_data.csv', index=False)
```
