```python
%pylab inline
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file = "clean_data/clean_victimization_data.csv"
df = pd.read_csv(data_file)
pd.set_option('display.max_columns', 50)
```

```python
s = (df.pivot_table(
        index='Year', columns='Type of crime', values = df.groupby('Type of crime').count(), aggfunc='sum'))
s.plot(kind='bar', stacked=True)
plt.show()
```
